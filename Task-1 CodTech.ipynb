# Training Linear Regression model

# Split the data into features(x) and target(y)

```python
X=df[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax','ptratio', 'b','lstat']].values
y=df[['medv']].values
```

# Train Test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.4,random_state=42)
```

# Training model using Linear Regression

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

# Make predictions

```python
predictions=model.predict(X_test)
```

# Evaluate model

```python
from sklearn import metrics
```

```python
print("Mean squared error:",metrics.mean_squared_error(y_test,predictions))
print("r2_score:",metrics.r2_score(y_test,predictions))
```

# Visualizing predictions from model

```python
sns.distplot((y_test-predictions),bins=50)
```

```python
plt.scatter(y_test,predictions,marker='o',edgecolors='black',c='blue')
plt.xlabel("Actual values")
plt.ylabel("Predicted values")
plt.title("Scatter plot of actual and prediced values")
```
